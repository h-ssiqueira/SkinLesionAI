```python
import tensorflow as tf

convNeXt = tf.keras.applications.ConvNeXtBase(weights=None, include_top=False, classes=7, input_shape=(224,224,3))
ResNet50V2 = tf.keras.applications.ResNet50V2(weights=None, include_top=False, classes=7, input_shape=(224,224,3))
Xception = tf.keras.applications.Xception(weights=None, include_top=False, classes=7, input_shape=(299,299,3))
ResNet152V2 = tf.keras.applications.ResNet152V2(weights=None, include_top=False, classes=7, input_shape=(224,224,3))
ResNet50 = tf.keras.applications.ResNet50(weights=None, include_top=False, classes=7, input_shape=(224,224,3))
```

```python
print('convNeXt: ', len(convNeXt.layers))
print('ResNet50V2: ', len(ResNet50V2.layers))
print('Xception: ', len(Xception.layers))
print('ResNet152V2: ', len(ResNet152V2.layers))
print('ResNet50: ', len(ResNet50.layers))
```

# ConvNeXt

```python
convNeXt.summary()
```

```python
for layer in convNeXt.layers:
    layer.trainable = False
prediction = tf.keras.layers.Dense(7, activation='softmax')(tf.keras.layers.Flatten()(convNeXt.output))

convNeXt = tf.keras.Model(inputs=convNeXt.input, outputs=prediction)
convNeXt.summary()
```
